### Section 1
First we will use BTC-USD five minute bars 
Download 2100 BTC-USD five-minute bars from GDAX, starting at 9am UTC 2017.12.01. Note that, as of early January 2018,
GDAX does not allow more than 350 items per request, and that there is also a limit on the number of requests per second.
You may write your own code, or use the following code in Python3. This code worked as of 2018.01.09, but GDAX may
change its specifications at any time and has done so in the past. Please check at https://docs.gdax.com/#get-trades.

```python
import pandas as pd
import numpy as np
# ========
from datetime import *
import time
from operator import itemgetter
import requests
REST_API_ENDPOINT_URL = 'https://api.gdax.com'
def get_historic_rates(product_id, start, end, granularity):
    r = requests.get(f'{REST_API_ENDPOINT_URL}/products/{product_id}/candles',
                     params={'start': start.isoformat(),
                             'end': end.isoformat(),
                             'granularity': str(granularity)})
    r.raise_for_status()
    return r.json()
```

```python
granularity = 300 # Five minutes
num_points = 2100
num_points_per_request = 300 # 350 / 5 minutes
begin_time = datetime(2017, 12, 1, 9)
end_time = begin_time + timedelta(seconds=granularity*num_points)
request_start = begin_time
rts = []

while request_start < end_time:
    request_end = request_start + timedelta(seconds=granularity*num_points_per_request)
    print('start = ', request_start, '; end = ', request_end, flush=True)
    tmp = sorted(get_historic_rates('BTC-USD', request_start, request_end, granularity), key=itemgetter(0))
    rts += tmp
    print('new lines: ', len(tmp), '; total lines: ', len(rts), flush=True)
    request_start = request_end
    time.sleep(1)
```

```python
# [time, low, high, open, close, volume]
rawdf = pd.DataFrame(rts, columns=['time','low','high','open','close','volume'])
btcdf = rawdf.copy()
btcdf['datetime'] = btcdf.time.apply(datetime.fromtimestamp)
```

### Section 2
Note that some of the bars will be empty, because the exchange was closed---see https://status.gdax.com/.
"It's up to you to decide how to deal with the missing data."

```python
# Find how many time periods have jumps greater than 300, implying gaps.
# For each time period, do a linear interpolation.
# Can do linear interpolation for marginal improvement in precision.
```

```python
def fill_missing(msg_idx):
    global btcdf
    current_time = btcdf.time.iloc[msg_idx-1] + granularity
    end_time = btcdf.time.iloc[msg_idx]
    
    first_half = btcdf.iloc[:msg_idx,:]
    second_half = btcdf.iloc[msg_idx:,:]

    to_insert = []
    while current_time < end_time:
        current_datetime = datetime.fromtimestamp(current_time)
        to_insert.append({
            'time':   current_time,
            'low':      np.nan,
            'high':     np.nan,
            'open':     np.nan,
            'close':    np.nan,
            'volume':   np.nan,
            'datetime': current_datetime
        })
        print("Filling in bar at {}: {}".format(current_time, current_datetime))
        current_time += granularity
        
     
    missing_rows = pd.DataFrame(to_insert, columns=btcdf.columns) 

    btcdf = pd.concat([first_half, missing_rows , second_half]).reset_index(drop=True) 
```

```python
# fill_gaps = False

# if fill_gaps:
msg_idxs = np.where(btcdf.time - btcdf.shift().time != 300)[0]
if len(msg_idxs)>1: #Check if missing_ind contains indices other than 0.
    for i in msg_idxs[1:]: # skip 0
        fill_missing(i)
        
btcdf.interpolate(method='linear', inplace=True); # Can change method to 'polynomial' or others.
```

### Section 4
Implement the bootstrapping procedure described in Appendix C of the paper.

*From Politis and Romano (1994)

```python
btcdf.shape
```

"Stationary Bootstrap is a resampling technique that can be applied to strictly stationary and weakly dependent 
  time series to generate a pseudo-time series that is stationary."
  
In general, a large value of q is appropriate for data with little dependence, whereas a small value of q is appropriate for data that exhibit more dependence.

The paper [Automatic Block-Length Selection for the Dependent Bootstrap](https://pdfs.semanticscholar.org/37ba/fa45629eec1cd713d6df954818f558f6d747.pdf), by Politis and White (2003), presents a pragmatic selection of a block size for bootstrapping. For a stationary bootstrap of $B = 800$ samples with data having an autocorrelation of $\rho=-0.4$, they recommend a block size of around 4.44, whereas one with $B = 200$ they recommend a block size of 2.7.  For $\rho = 0.1$ they recommend 1.3 if $B=200$ and 2.0 if $B=800$.



```python
btcdf.returns.autocorr()
```

For our purposes, we ballpark an appropriate block size of 3, making $q = 1/3$ as this stationary bootstrap follows a geometric distribution so the expected block length is $1/q$. For this, we choose $B=500$, the same number of samples used in the paper.

```python
# Stationary Bootstrap is a resampling technique that can be applied to strictly stationary and weakly dependent 
#  time series to generate a pseudo-time series that is stationary.

# More literature 


"""Implemented by index instead of value."""

timeindex = range(btcdf.shape[0])

R = 0
T = timeindex[-1]

def get_smoothing_param():
    q = np.random.random() #Produces [0, 1)
    q = 1 if q == 0 else q # Produces (0, 1]
    return q

def get_theta(timeindex):
     return np.random.choice(timeindex)

def generate_block(R, T, q=None, verbose=0):
    if not q:
        q = get_smoothing_param()
        if verbose:
            print("Smoothing Parameter: {:.3f}".format(q))
        
    t = R+1
    theta_t = [get_theta(timeindex)]
    while t < T:
        t += 1
        U = np.random.random()
        if U < q:
            theta_t.append(get_theta(timeindex))
        else:
            next_theta = theta_t[-1] + 1
            if next_theta > T:
                next_theta = R
            theta_t.append(next_theta)
            
    if verbose==2:
        print("Length of bootstrap sample: {} indexes.".format(len(theta_t)))
    return theta_t

def generate_blocks(B, R, T, q = None, verbose=0):
    blocks = []
    for i in range(B):
        if verbose==1:
            if i%50==0:
                print("Processing block {} of {}".format(i+1, B))
        if verbose==2:
            print("Processing block {} of {}".format(i+1, B))
        blocks.append(generate_block(R, T, q, verbose))
    print("Done processing {} blocks!".format(i+1))
    return blocks

q = 1/3
test_blocks = generate_blocks(1, R, T, q, 1)
```

Implement any four strategies from among the ones described in Section III and Appendix A of the paper. The choice of what

```python

features = {
    'num_strategies': 4,
    'mode': ['simple','simple','obv','obv'],
    'window': [10, 15, 25, 55],
    'b_filter': [0.005, 0.008, 0.003, 0.001],
    'd_delay' : [2, 4, 2, 3],
    'c_delay': [25, 35, 35, 45]
}

def band_filter(row, col_name, b_filter = 0.001):
    """1 if greater than close+filter, -1 if less than close-filter"""
    if row.close > row[col_name]*(1+b_filter):
        return 1
    if row.close < row[col_name]*(1-b_filter):
        return -1
    return 0

def days_filter(df, d_delay = 2, c_delay = 35):
    """Given a dataframe with a signal, zero out signal that has been on for either less than d_day
       or more than or equal to c_day."""
    counter = 0
    prev_signal = df.loc[0, 'signal']
    for i, row in df.iterrows():
        row = row.signal
        # Only increase counter if signal is not 0,
        #  and signal did not swing from -1 to 1 or viceversa. 
        #  Else reset to 0.
        
        if (row != 0) and (prev_signal != -(row)):
            counter += 1
        else:
            counter = 0
            
        # Only trade when signal has been on for d_days or it expired.
        if counter < d_delay:
            df.loc[i,'signal'] = 0
        if counter >= c_delay:
            df.loc[i,'signal'] = 0
            
        prev_signal = row
    
    return df.signal

def obv_map(bool_val):
    """Remaps boolean values. True => 1, False => -1"""
    return 1 if bool_val else -1

def get_MA_profit(btcdf, window, b_filter, d_delay, c_delay, mode='simple'):
    col_name = "MA_{}_{}".format(window, mode)
    profit_col_name = 'profit_{}'.format(col_name)
    ma_df = btcdf[['time', 'close', 'volume','datetime', 'returns']].copy()
    
    if mode=='obv':
        close_px = btcdf.close.copy()
        ma_df.loc[:,'obv'] = (ma_df.volume*((close_px - close_px.shift())>0)
                                   .map(obv_map)).cumsum()
        ma_df.loc[:,col_name] = ma_df.obv.rolling(window=window).mean()
        
    elif mode == 'simple': 
        ma_df.loc[:,col_name] = ma_df.close.rolling(window=window).mean()
        
    ma_df.loc[:, 'signal'] = ma_df.apply(lambda x: band_filter(x, col_name, b_filter), axis=1)
    ma_df.loc[:, 'signal'] = days_filter(ma_df, d_delay, c_delay)
    ma_df.loc[:, 'profit'] = (ma_df.signal*ma_df.returns).cumsum()
    ma_df.rename(columns={'profit' : profit_col_name}, inplace=True)
    not_cumsum_name = 'not_cumsum_{}'.format(profit_col_name)
    ma_df.loc[:, not_cumsum_name ] = ma_df.signal*ma_df.returns
    return ma_df[[col_name, profit_col_name, not_cumsum_name]]

def get_MA_profit_from_feats(btcdf, features):
    profits = [btcdf]
    for i in range(features.get('num_strategies')):
        window_size = features.get('window')[i]
        mode = features.get('mode')[i]
        print("Processing {} {} moving average".format(window_size, mode))
        profits.append(get_MA_profit(btcdf, 
                           window_size,
                           features.get('b_filter')[i],
                           features.get('d_delay')[i],
                           features.get('c_delay')[i],
                           mode
                        ))
    return pd.concat(profits, axis=1)

```

```python
# Buy and Hold Strategy
# Mean return is interpreted as total return if, instead of annualizing, we take the entire time period as our 
#  sample space.  Additionally, since they're all coming from the same instrument and only used to compare 
#  to each other, we don't take the log of the returns. Portfolio standard deviation is taken as the deviation of
#  our cumulative returns to date.


RF_RATE = 0 # Set to 0 as risk free rate over six days is very small.


def calc_return_pctg(series):
    series.dropna(inplace=True)
    return (series.iloc[-1]-series.iloc[0]) / series.iloc[0]

def calc_strat_pctg(c_return, close_px):
    return c_return.iloc[-1] / close_px.iloc[0]

def calc_sharpe(c_returns, rf_rate=RF_RATE):
    returns = c_returns.iloc[-1]
    return_std = c_returns.std()
    sharpe = (returns - rf_rate) / return_std
    return sharpe

buy_and_hold_return = calc_return_pctg(btcdf.close)
buy_and_hold_sharpe = calc_sharpe(btcdf.c_returns)

simple10_return = calc_strat_pctg(ma_df.profit_MA_10_simple, ma_df.close)
simple10_sharpe = calc_sharpe(ma_df.profit_MA_10_simple)

simple15_return = calc_strat_pctg(ma_df.profit_MA_15_simple, ma_df.close)
simple15_sharpe = calc_sharpe(ma_df.profit_MA_15_simple)

obv25_return = calc_strat_pctg(ma_df.profit_MA_25_obv, ma_df.close)
obv25_sharpe = calc_sharpe(ma_df.profit_MA_25_obv)

obv55_return = calc_strat_pctg(ma_df.profit_MA_55_obv, ma_df.close)
obv55_sharpe = calc_sharpe(ma_df.profit_MA_55_obv)

print("Buy and hold strategy earns a total of ${:.2f} ({:.2f}%) return with {:.2f} Sharpe ratio.\n"
      .format(btcdf.c_returns.iloc[-1], 100*(buy_and_hold_return), buy_and_hold_sharpe))

print("Simple 10 MA earns a total of ${:.2f} ({:.2f}%) return with {:.2f} Sharpe ratio.\n"
      .format(ma_df.profit_MA_10_simple.iloc[-1], 100*(simple10_return), simple10_sharpe))

print("Simple 15 MA earns a total of ${:.2f} ({:.2f}%) return with {:.2f} Sharpe ratio.\n"
      .format(ma_df.profit_MA_15_simple.iloc[-1], 100*(simple15_return), simple15_sharpe))

print("OBV 25 ears earns a total of ${:.2f} ({:.2f}%) return with {:.2f} Sharpe ratio.\n"
      .format(ma_df.profit_MA_25_obv.iloc[-1], 100*(obv25_return), obv25_sharpe))

print("OBV 55 earns a total of ${:.2f} ({:.2f}%) return with {:.2f} Sharpe ratio.\n"
      .format(ma_df.profit_MA_55_obv.iloc[-1], 100*(obv55_return), obv55_sharpe))
```

#### White's Reality Check

Steps:
    - 1. Compute benchmark as percent returns (As long as the type of return is uniform accross strategies, check should still compute).
    - 2. Compute excess return per day for each strategy
    - 3. Compute mean excess return per bootstrap and total. (B = 500, with q in [ 0.01, 0.1, 0.5] ).
    - 4. Compute Vs per strategy
    - 5. Find max Vs of all strategies
    - 6. Calculate p =  1 - # Vs bootstrapped values higher than total Vs divided by N.
    
    
The paper uses B = 500, so we set it as 500 as well.

```python
B = 500
R = 55 # Latest strategy start
T = 2100 # Number of points
q = 1/3

blocks = generate_blocks(B, R, T, q, 1)
```

```python
def compute_excess(strategies, benchmark):
    return strategies.subtract(benchmark, axis=0)

def compute_mean(strategy):
    return strategy.mean()

def compute_bootstrapped_means(strategy, blocks):
    return [strategy.loc[block].mean() for block in blocks]

def compute_v_total(strategy):
    n = len(strategy)
    mean = compute_mean(strategy)
    return (n**0.5)*(mean)

def compute_v_bootstrap(strategy, blocks):
    n = len(strategy)
    mean = compute_mean(strategy)
    bootstrapped_means = compute_bootstrapped_means(strategy, blocks)
    return (n**2)*(bootstrapped_means - mean)

def compute_max_v_total(strategies):
    V = -np.inf
    for column in strategies:
        V = max(V, compute_v_total(strategies[column]))
    return V

def compute_max_v_bootstrap(strategies, blocks):
    V = -np.inf
    for column in strategies:
        V = np.maximum(V, compute_v_bootstrap(strategies[column], blocks))
    return V

def compute_p(strategies, blocks):
    """ 
    M / N where M is better than Vs and N is the number of bootstrapped samples
    """
    max_total = compute_max_v_total(strategies)
    max_bootstrap = compute_max_v_bootstrap(strategies, blocks)
    better_than_vs = (max_bootstrap < max_total).sum()
    return 1 - better_than_vs / len(blocks)
```
